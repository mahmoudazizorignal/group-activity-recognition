```python
import torch
from helpers.config import get_settings
from controllers import AnnotationController
```

```python
settings = get_settings()
torch.manual_seed(settings.SEED)
torch.cuda.manual_seed(settings.SEED)
annotations = AnnotationController.get_annotations(settings=settings)
```

```python
cnt1, cnt2 = 0, 0

for video in annotations.keys():
    for frame in annotations[video].keys():
        cnt = 0
        for i, player in enumerate(annotations[video][frame]["players"]):
            if len(player) == 0:
                cnt += 1
        if cnt:
            cnt1 += 1
        cnt2 += 1
        
print(f"{cnt1}/{cnt2}")
```
